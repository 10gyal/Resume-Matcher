```python
nlp = spacy.load('en_core_web_lg')
```

```python
import pandas as pd
df = pd.read_json('../Dataset/Resume_1.json')
```

```python
cols = list(df.columns)
```

```python
resume1 = df['Person 1'][0]
```

```python
doc = nlp(resume1)
```

```python
# displacy.render(doc, style="ent")
```

```python
def display():
    for column in cols:
        resume = df[column][0]
        doc = nlp(resume)
        displacy.render(doc, style="ent")
        print("\n","="*100,"\n")
        
```

```python
display()
```
